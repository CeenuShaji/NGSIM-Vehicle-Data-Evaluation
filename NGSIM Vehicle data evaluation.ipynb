```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Load the data (replace 'your_file.csv' with your actual file path)
file_path = 'PeachTree_Next_Generation_Simulation__NGSIM__Vehicle_Trajectories_and_Supporting_Data (1).csv'
data = pd.read_csv(file_path)

# Select a random Vehicle_ID
random_vehicle_id = random.choice(data['Vehicle_ID'].unique())

# Filter data for the randomly selected vehicle
vehicle_data = data[data['Vehicle_ID'] == random_vehicle_id]

# Create a plot
plt.figure(figsize=(10, 6))

# Plot Global_X vs Global_Y for the randomly selected vehicle
plt.scatter(vehicle_data['Global_X'], vehicle_data['Global_Y'], label=f'Vehicle {random_vehicle_id}')

# Adding title and labels
plt.title(f'Global X vs Global Y for Vehicle ID {random_vehicle_id}')
plt.xlabel('Global X')
plt.ylabel('Global Y')
plt.legend()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random


file_path = 'PeachTree_Next_Generation_Simulation__NGSIM__Vehicle_Trajectories_and_Supporting_Data (1).csv'
data = pd.read_csv(file_path)

# Select a random Vehicle_ID
random_vehicle_id = random.choice(data['Vehicle_ID'].unique())

# Filter data for the randomly selected vehicle
vehicle_data = data[data['Vehicle_ID'] == random_vehicle_id]

# Create a plot
plt.figure(figsize=(10, 6))

# Iterate through each unique Int_ID for this vehicle
for int_id in vehicle_data['Int_ID'].unique():
    # Filter data for the current Int_ID
    int_id_data = vehicle_data[vehicle_data['Int_ID'] == int_id]
    
    # Plot Global_X vs Global_Y for this Int_ID
    plt.scatter(int_id_data['Global_X'], int_id_data['Global_Y'], label=f'Int_ID {int_id}')

# Adding title and labels
plt.title(f'Global X vs Global Y for Vehicle ID {random_vehicle_id} (Different Int_IDs)')
plt.xlabel('Global X')
plt.ylabel('Global Y')
plt.legend()

# Display the plot
plt.show()
```
